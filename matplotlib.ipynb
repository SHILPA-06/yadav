```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y, 'c')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('string')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x, y, 'r--')
plt.subplot(1,2,2)
plt.plot(y, x, 'g*-')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y, 'p')
axes.set_xlabel('x label')
axes.set_ylabel('y label')
axes.set_title('data')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x, y, 'b')
axes1.set_xlabel('x_label_axes2')
axes1.set_ylabel('y_label_axes2')
axes1.set_title('axes 1 tile')

axes2.plot(y, x, 'r')
axes2.set_xlabel('x_label_axes2')
axes2.set_ylabel('y_label_axes2')
axes2.set_title('axes 2 title')
```

```python
fig, axes = plt.subplots(figsize=(12,2.5))

axes.plot(x,y, 'r')
axes.plot('x')
axes.plot('y')
axes.set_title('title')
```
